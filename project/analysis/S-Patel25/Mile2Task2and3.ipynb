```python
import sys
import pandas as pd
import seaborn as sns
sys.path.append('..')
sns.set_style("darkgrid")
import matplotlib as plt
```

```python
from scripts import project_functions as pf
top4 = pf.load_process('C:/Users/patel/COSC301/Group22-Project/project-group22-project/data/raw/season-1819_csv.csv')
print(top4)
```

```python
print(top4.info())
print(top4.describe().T)
```

```python
barcaHome = top4[top4['HomeTeam'] == 'Barcelona']
barcaAway = top4[top4['AwayTeam'] == 'Barcelona']

barc = pd.concat([barcaHome, barcaAway])
barc

sns.lineplot(x = "FTR", y = "HF", data = barc).set(title = "Barcelona total fouls compared with the result of the game", xLabel = "Game Decision (H = Home Win, D = Draw, A = Away Win)", yLabel = "Home Fouls")
sns.lineplot(x = "FTR", y = "AF", data = barc)
```

# This lineplot shows that Barcelona tends to win more if they foul less in the game regardless if they are playing in Barcelona or elsewhere (Blue = Barc at home, Orange = Barc at away).

```python
realMHome = top4[top4['HomeTeam'] == 'Real Madrid']
realMAway = top4[top4['AwayTeam'] == 'Real Madrid']

realM = pd.concat([realMHome, realMAway])
realM

sns.lineplot(x = "FTR", y = "HST", color = "Green", data = realM).set(title = "Real Madrid shots on target compared with games won", xLabel = "Game Decision (H = Home Win, D = Draw, A = Away Win)", yLabel = "Shots on target")
sns.lineplot(x = "FTR", y = "AST", color = "Purple", data = realM)
```

# This plot shows that Real Madrid wins siginifcantly more games when they have more shots on target, and it does not affect them if they are at home (green) or away (purple)

```python
valenciaHome = top4[top4['HomeTeam'] == 'Valencia']
valenciaAway = top4[top4['AwayTeam'] == 'Valencia']

val = pd.concat([valenciaHome, valenciaAway])
val

sns.lineplot(x = "FTHG", y = "HC", color = "Black", data = val).set(title = "Valencia total corner shots compared with goals scored", xLabel = "Goals Scored", yLabel = "Corners Shot")
sns.lineplot(x = "FTAG", y = "AC", color = "Blue", data = val)
```

# Plot shows how Valencia not only scores more with more corners (more goal oppurtunities), but also that they shoot more corners and goals at away games which is a different trend compared to the other 2 teams
