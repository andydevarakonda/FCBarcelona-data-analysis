```python
Real_Madrid_h = df[df["HomeTeam"] == "Real Madrid"]
Real_Madrid_h
```

```python
Real_Madrid_away = df[df["AwayTeam"] == "Real Madrid"]
Real_Madrid_away
```

```python
RM = pd.concat([Real_Madrid_h,Real_Madrid_away])
RM
```

```python
diff_HS_RM = Real_Madrid_h["HS"]-Real_Madrid_h["AS"]
diff_AS_RM = Real_Madrid_away["AS"]-Real_Madrid_away["HS"]
```

```python
diff_HS_barca = df[df["HomeTeam"] == "Barcelona"]["HS"]-df[df["HomeTeam"] == "Barcelona"]["AS"]
diff_AS_barca = df[df["AwayTeam"] == "Barcelona"]["AS"]-df[df["AwayTeam"] == "Barcelona"]["HS"]
```

```python
diff_HS_Ath_Mardid = df[df["HomeTeam"] == "Ath Madrid"]["HS"]-df[df["HomeTeam"] == "Ath Madrid"]["AS"]
diff_AS_Ath_Madrid = df[df["AwayTeam"] == "Ath Madrid"]["AS"]-df[df["AwayTeam"] == "Ath Madrid"]["HS"]
```

```python
diff_HS_Valencia = df[df["HomeTeam"] == "Valencia"]["HS"]-df[df["HomeTeam"] == "Valencia"]["AS"]
diff_AS_Valencia = df[df["AwayTeam"] == "Valencia"]["AS"]-df[df["AwayTeam"] == "Valencia"]["HS"]
```

```python
diff_RM = pd.concat([diff_HS_RM,diff_AS_RM])
diff_RM.reset_index(drop=True, inplace=True)
diff_barca = pd.concat([diff_HS_barca,diff_AS_barca])
diff_barca.reset_index(drop=True, inplace=True)
diff_Ath_Madrid = pd.concat([diff_HS_Ath_Mardid,diff_AS_Ath_Madrid])
diff_Ath_Madrid.reset_index(drop=True, inplace=True)
diff_valencia = pd.concat([diff_HS_Valencia,diff_AS_Valencia])
diff_valencia.reset_index(drop=True, inplace=True)


```

```python
df1 = pd.DataFrame();
df1.insert(0, "Difference Real Madrid",diff_RM )
df1.insert(1,"Difference Barca",diff_barca)
df1.insert(2,"Difference Ath Madrid ",diff_Ath_Madrid)
df1.insert(3," Difference Valencia ",diff_valencia )
df1
```

```python
sns.displot(data = df1,kde=True)
```
