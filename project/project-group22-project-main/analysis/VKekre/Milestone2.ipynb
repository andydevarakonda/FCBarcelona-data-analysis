```python
import pandas as pd

def load_process(path_to_csv):
    file = pd.read_csv(path_to_csv)
    df= pd.DataFrame(file)
    updt_df = (pd.read_csv(path_to_csv).iloc[0:len(df), 1:22])

    barca_homestadium= (updt_df[updt_df['HomeTeam']== 'Barcelona'])
    barca_other= (updt_df[updt_df['AwayTeam']== 'Barcelona'])
    barca_total = (pd.concat([barca_homestadium,barca_other]))


    realmadrid_homestadium= updt_df[updt_df['HomeTeam']== 'Real Madrid']
    realmadrid_other= updt_df[updt_df['AwayTeam']== 'Real Madrid']
    realmadrid_total = pd.concat([realmadrid_homestadium,realmadrid_other])


    atletico_homestadium= updt_df[updt_df['HomeTeam']== 'Ath Madrid']
    atletico_other= updt_df[updt_df['AwayTeam']== 'Ath Madrid']
    atletico_total = pd.concat([atletico_homestadium, atletico_other])


    valencia_homestadium= updt_df[updt_df['HomeTeam']== 'Valencia']
    valencia_other= updt_df[updt_df['AwayTeam']== 'Valencia']
    valencia_total = pd.concat([ valencia_homestadium,valencia_other])

    top_4 = (pd.concat([barca_total, realmadrid_total, atletico_total, valencia_total]))
    return top_4
  
```

```python
load_process("season-1819_csv.csv")
```
