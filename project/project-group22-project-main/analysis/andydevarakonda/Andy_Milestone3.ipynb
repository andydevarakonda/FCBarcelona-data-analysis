```python
barca_home = top_4[top_4["HomeTeam"] == "Barcelona"]
barca_home
```

```python
barca_away = top_4[top_4["AwayTeam"] == "Barcelona"]
barca_away
```

```python
barcelona = pd.concat([barca_home, barca_away])
barcelona
```

```python
goal_diffh_barca = barca_home["FTHG"]-barca_home["FTAG"]
goal_diffa_barca = barca_away["FTAG"]-barca_away["FTHG"]

goal_diffh_rma = top_4[top_4["HomeTeam"] == "Real Madrid"]["FTHG"]-top_4[top_4["HomeTeam"] == "Real Madrid"]["FTAG"]
goal_diffa_rma = top_4[top_4["AwayTeam"] == "Real Madrid"]["FTAG"]-top_4[top_4["AwayTeam"] == "Real Madrid"]["FTHG"]

goal_diffh_atl =top_4[top_4["HomeTeam"] == "Ath Madrid"]["FTHG"]-top_4[top_4["HomeTeam"] == "Ath Madrid"]["FTAG"]
goal_diffa_atl = top_4[top_4["AwayTeam"] == "Ath Madrid"]["FTAG"]-top_4[top_4["AwayTeam"] == "Ath Madrid"]["FTHG"]

goal_diffh_val = top_4[top_4["HomeTeam"] == "Valencia"]["FTHG"]-top_4[top_4["HomeTeam"] == "Valencia"]["FTAG"]
goal_diffa_val = top_4[top_4["AwayTeam"] == "Valencia"]["FTAG"]-top_4[top_4["AwayTeam"] == "Valencia"]["FTHG"]



goal_diff_barca = pd.concat([goal_diffh_barca,goal_diffa_barca])
goal_diff_barca.reset_index(drop=True, inplace=True)

goal_diff_rma = pd.concat([goal_diffh_rma,goal_diffa_rma])
goal_diff_rma.reset_index(drop=True, inplace=True)

goal_diff_atl = pd.concat([goal_diffh_atl,goal_diffa_atl])
goal_diff_atl.reset_index(drop=True, inplace=True)

goal_diff_val = pd.concat([goal_diffh_val,goal_diffa_val])
goal_diff_val.reset_index(drop=True, inplace=True)



goal_diff = pd.DataFrame();


goal_diff.insert(0, "Barca Goal Diff", goal_diff_barca)
goal_diff.insert(1,"Real Madrid Goal Diff", goal_diff_rma)
goal_diff.insert(2,"Ath Madrid Goal Diff", goal_diff_atl)
goal_diff.insert(3," Valencia Goal Diff", goal_diff_val)
goal_diff


homeshots_barca = barca_home["HST"]
awayshots_barca = barca_away["AST"]

homeshots_rma= top_4[top_4["HomeTeam"] == "Real Madrid"]["HST"]
awayshots_rma = top_4[top_4["AwayTeam"] == "Real Madrid"]["AST"]

homeshots_atl =top_4[top_4["HomeTeam"] == "Ath Madrid"]["HST"]
awayshots_atl = top_4[top_4["AwayTeam"] == "Ath Madrid"]["AST"]

homeshots_val = top_4[top_4["HomeTeam"] == "Valencia"]["HST"]
awayshots_val = top_4[top_4["AwayTeam"] == "Valencia"]["AST"]

homeshots_barca_total = pd.concat([homeshots_barca,awayshots_barca])
homeshots_barca_total.reset_index(drop=True, inplace=True)

homeshots_rma_total = pd.concat([homeshots_rma,awayshots_rma])
homeshots_rma_total.reset_index(drop=True, inplace=True)

homeshots_atl_total = pd.concat([homeshots_atl,awayshots_atl])
homeshots_atl_total.reset_index(drop=True, inplace=True)

homeshots_val_total = pd.concat([homeshots_val,awayshots_val])
homeshots_val_total.reset_index(drop=True, inplace=True)


goal_diff.insert(4, "Barca Shots on target", homeshots_barca_total)
goal_diff.insert(5,"Real Madrid Shots on target", homeshots_rma_total)
goal_diff.insert(6,"Ath Madrid Shots on target", homeshots_atl_total)
goal_diff.insert(7," Valencia Shots on target", homeshots_val_total)
goal_diff
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
sns.barplot(data=goal_diff,ax = ax)



```

# The above graph showcases the barplots between the goal difference and the shots on target for the top 4 teams
